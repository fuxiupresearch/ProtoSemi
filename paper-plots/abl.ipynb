```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = {
    'font.sans-serif': 'Arial Unicode MS',
    'font.size': 16, 
    'figure.figsize': '12.5, 5',   
}
plt.rcParams.update(params)
```

```python
# set width of bar
barWidth = 0.15
unit_len = 0.9
fig = plt.subplots(dpi=500)

# set height of bar
y1 = [95.03, 95.48, 95.48, 95.67, 92.97, 67.73]
y2 = [94.46, 95.10, 94.95, 95.01, 92.32, 66.50]
y3 = [88.33, 86.44, 86.43, 86.20, 79.93, 52.97]


# Set position of bar on X axis
br1 = np.arange(len(y1)) * unit_len + 0.3
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
# print(br1, br2)

plt.bar(br1, y1, color='#FA7F6F', width=barWidth, edgecolor='black', label='Proto-semi')

plt.bar(br2, y2, color='#FFBE7A', width=barWidth, edgecolor='black', label='Proto-semi-w/o-repar')

plt.bar(br3, y3, color='#8ECFC9', width=barWidth,  edgecolor='black',label='Proto-semi-w/o-semi')


plt.xlim((-0.1, 5.5))
plt.ylim((50, 100))

# Adding Xticks
plt.xlabel('Dataset', fontweight='bold')
plt.ylabel('Accuracy', fontweight='bold')
plt.xticks([r * unit_len + 3.0 * barWidth for r in range(len(y1))],
           ['aggre', 'rand1', 'rand2', 'rand3', 'worst', 'noisy'])

# fig.legend(ncol=4,loc='upper center', bbox_to_anchor=(2.4,1.2),bbox_transform=ax1.transAxes)
plt.legend(ncol=1, loc='upper right', fontsize=12)
plt.savefig('abl.pdf', bbox_inches='tight')
plt.show()
```
